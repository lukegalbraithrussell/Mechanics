# Packages

```python
from scipy import *
from pylab import *
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set() 
```

# Functions and Constants

```python
#differential equation
def rhs(x,t):
    x0dot = x[1]
    x1dot = - d*x[1] - alpha*x[0] - beta*x[0]*x[0]*x[0] + gamma*cos((w0)*t)
    return [x0dot, x1dot]

#timespan
tmax = 2000
t = linspace(0, tmax, 10000)

#Parameters
w0 = 1.4
gamma = .34
alpha = -1
beta = 1
d =.1

init_x = 0
init_v = 0.0
init_cond = [init_x,init_v]
ces = odeint(rhs, init_cond, t)
```

# Basic Plots

```python
#Plot 1
plt.xlabel("Time, t")
plt.ylabel("Position, v")
plt.plot(t[-500:],ces[-500:,0],'-b')
plt.title("Time t vs. Position x for a Duffing oscillator.")
plt.show()

#Plot 2
plt.xlabel("Position, x")
plt.ylabel("Velocity, v")
plt.plot(ces[-250:,0],ces[-250:,1],'-b')
plt.title("Position x vs. Velocity v for a Duffing oscillator.")
plt.show()

#Plot 3
plt.xlabel("Time, t")
plt.ylabel("Velocity, v")
plt.plot(t[-500:],ces[-500:,1],'-b')
plt.title("Time t vs. Velocity v for a Duffing oscillator.")
plt.show()
```

# Poincare Map

```python
gamma = .35
pi = 3.1415
t = np.arange(0, 10000, 2*pi/w0)

init_x = 0
init_v = 0.0
init_cond = [init_x,init_v]
ces = odeint(rhs, init_cond, t)

plt.xlabel("Position, x")
plt.ylabel("Velocity, v")
plt.scatter(ces[:,0],ces[:,1], s=15, alpha=.50)
plt.title("Position x vs. Velocity v for a Duffing oscillator.")
plt.show()
```

# Bifurcation Diagram

```python
t = np.arange(0, 6000, 2*pi/w0)
g_array = np.arange(.3,.35,.00025)

init_x = 0
init_v = 0.0
init_cond = [init_x,init_v]

glist = []
nlist = []

for g in g_array:
    def rhs(x,t):
        x0dot = x[1]
        x1dot = - d*x[1] - alpha*x[0] - beta*x[0]*x[0]*x[0] + g*cos((w0)*t)
        return [x0dot, x1dot]
    ces = odeint(rhs, init_cond, t)
    numbers = ces[-200:,0]
    glist.extend(g*np.ones(len(numbers)))
    nlist.extend(numbers)
    
plt.xlabel("Driving Force, Gamma")
plt.ylabel("Position, x")
plt.scatter(glist, nlist, s=3, alpha =.50)
plt.title("Driving Force vs Position")
plt.show()
```
